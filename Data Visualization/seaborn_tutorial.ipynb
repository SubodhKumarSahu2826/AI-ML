```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
print(type(tips))

print(tips.head())
```

```python
sns.set_theme()


sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    style="smoker"
)
```

```python
#Scatter plot

tips=sns.load_dataset("tips")
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
#line plot
sns.lineplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
#line plot
flights=sns.load_dataset("flights")
sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```

```python
print(tips.head())
```

```python
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
#box plot
sns.boxplot(
    data=tips,
    x="day",
    y="tip"
)
```

```python
#Historgrams

penguins=sns.load_dataset("penguins")

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)
```

```python
penguins.head()
```

```python
#Matrix Plots(Heat maps)

flights.head()
```

```python
flights_pivot=flights.pivot(index="month", columns="year", values="passengers")

sns.heatmap(
    flights_pivot,cmap="coolwarm"
)
plt.title("Passenger Heatmaps")
```

```python
fig, ax=plt.subplots()

sns.lineplot(
    data=tips,x="day",
    y="total_bill",
    hue="sex",
    ax=ax
)

ax.set_title("Average bill value for each day")
ax.set_xlabel("Days")
ax.set_ylabel("total Bill Amount")
```
